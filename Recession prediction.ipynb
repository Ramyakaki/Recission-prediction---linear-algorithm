```python
import matplotlib.pyplot as plt 
import seaborn as sns

# Using pairplot we'll visualize the data for correlation
sns.pairplot(data, x_vars=['pop', 'emp',], y_vars='rdana', height=4, aspect=1, kind='scatter')
plt.show()
```

```python
df = pd.DataFrame(data)

import numpy as np
from sklearn.model_selection import train_test_split

X = np.asarray(df[['pop', 'emp']])
Y = np.asarray(df['rdana'])
```

```python
from sklearn.model_selection import train_test_split
X_train_lm, X_test_lm, y_train_lm, y_test_lm = train_test_split(X, Y, train_size = 0.7, test_size = 0.3, random_state = 100)
```

```python
from sklearn.linear_model import LinearRegression

# Creating an object of Linear Regression
lm = LinearRegression()

# Fit the model using .fit() method
lm.fit(X_train_lm, y_train_lm)
```

```python
y_pred=lm.predict(X_test_lm)
print(lm.coef_)
```

```python
# Intercept value
print("Intercept :",lm.intercept_)

# Slope value
print('Slope :',lm.coef_)
```

```python
from sklearn.metrics import r2_score
r2_score(y_test_lm,y_pred)
```

```python
p= sns.regplot(x=y_test_lm, y= y_pred,ci=None,scatter_kws={"color": "blue"},line_kws={"color":"red"})
p.set_xlabel("actual_rdana",fontsize=10)
p.set_ylabel("predicted_rdana",fontsize=10)
```
